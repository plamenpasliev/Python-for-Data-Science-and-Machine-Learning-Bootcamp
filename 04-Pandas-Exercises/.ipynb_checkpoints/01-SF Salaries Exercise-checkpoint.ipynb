**What is the average BasePay ?**

```python
df.loc[df['TotalPayBenefits'] == df['TotalPayBenefits'].max()]
```

```python
df['BasePay'].mean()
```

** What is the highest amount of OvertimePay in the dataset ? **

```python
df['OvertimePay'].max()
```

** What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll). **

```python
df.loc[df['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

** How much does JOSEPH DRISCOLL make (including benefits)? **

```python
df.loc[df['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']
```

** What is the name of highest paid person (including benefits)?**

```python
df.loc[df['TotalPayBenefits'] == df['TotalPayBenefits'].max()]
```

** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
df.loc[df['TotalPayBenefits'] == df['TotalPayBenefits'].min()]
```

```python
df.groupby(by='JobTitle').describe()
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
df.groupby(by='Year').mean()['BasePay']
```

** How many unique job titles are there? **

```python
df['JobTitle'].nunique()
```

```python
df['JobTitle'].value_counts().head()
```

```python
df.groupby(by='JobTitle').agg(lambda x:x.count())
```

```python
jobcount = df[df['Year']==2013].groupby(by='JobTitle').count()
```

** What are the top 5 most common jobs? **

```python
df.groupby(by='JobTitle').count().sort_values('Id', ascending=False)['Id'].head()
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
len(jobcount[jobcount['Id'] == 1])
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
df['JobTitle'].apply(lambda x: 'Chief' in x).sum()
```

```python
df['JobTitle']
```
